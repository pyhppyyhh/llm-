```python
from google.colab import drive
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch

# Drive 마운트
drive.mount('/content/drive')

SAVE_DIR = "/content/drive/MyDrive/qwen3-14b-finetuned-final"

# 베이스 모델 로드
model_name = "Qwen/Qwen3-14B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

# LoRA 어댑터 얹기
model = PeftModel.from_pretrained(model, SAVE_DIR)
model.eval()

print("✅ 파인튜닝 모델 로드 완료!")
print(f"GPU 메모리: {torch.cuda.memory_allocated() / 1024**3:.1f}GB / 40GB")
```

```python
# ============================================================
# 셀 4 - 모델 테스트 (로드 확인용)
# ============================================================
messages = [{"role": "user", "content": "안녕! 자기소개 해봐"}]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True,
    enable_thinking=False   # Thinking 모드 OFF
)
inputs = tokenizer(text, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_new_tokens=200)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
# ============================================================
# 셀 5 - LoRA 설정
# ============================================================
from peft import LoraConfig, get_peft_model, TaskType

lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ],
    bias="none",
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python


# ============================================================
# 셀 7 - 한번에 학습
# ============================================================
from datasets import Dataset
from google.colab import drive
from trl import SFTTrainer, SFTConfig
from peft import LoraConfig, get_peft_model, TaskType
import re

# Google Drive 마운트
drive.mount('/content/drive')
SAVE_DIR = "/content/drive/MyDrive/qwen3-14b-finetuned"

SYSTEM_PROMPT = "당신은 재단 사이트의 업무를 안내하는 친절한 챗봇입니다. 정확하고 간결하게 답변하세요."

# LoRA 초기화 (매번 새로 초기화)
lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ],
    bias="none",
)
model_train = get_peft_model(model, lora_config)
model_train.print_trainable_parameters()

# cite 태그 제거
def clean_data(data):
    cleaned = []
    for item in data:
        instruction = re.sub(r'\[cite:.*?\]', '', item["instruction"]).strip()
        response = re.sub(r'\[cite:.*?\]', '', item["response"]).strip()
        cleaned.append({"instruction": instruction, "response": response})
    return cleaned

raw_data = clean_data(raw_data)

# 데이터 포맷
def format_chat(example):
    messages = [
        {"role": "system",    "content": SYSTEM_PROMPT},
        {"role": "user",      "content": example["instruction"]},
        {"role": "assistant", "content": example["response"]},
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False,
        enable_thinking=False
    )
    text = text.replace("<think>\n\n</think>\n\n", "")
    return {"text": text}

dataset = Dataset.from_list(raw_data)
dataset = dataset.map(format_chat, remove_columns=["instruction", "response"])
split = dataset.train_test_split(test_size=0.2, seed=42)

# ✏️ 데이터 수에 따라 epoch 조정
# 100개 이하 → 2
# 200~500개 → 3
# 500개 이상 → 3~5
EPOCHS = 3

sft_config = SFTConfig(
    output_dir=SAVE_DIR,
    num_train_epochs=EPOCHS,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=4,
    learning_rate=2e-4,
    lr_scheduler_type="cosine",
    warmup_steps=10,
    fp16=True,
    gradient_checkpointing=True,
    optim="paged_adamw_8bit",
    logging_steps=5,
    eval_strategy="steps",
    eval_steps=20,
    save_steps=20,
    save_total_limit=2,
    load_best_model_at_end=True,
    dataset_text_field="text",
)

train_dataset = split["train"]
eval_dataset  = split["test"]

trainer = SFTTrainer(
    model=model_train,
    args=sft_config,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    processing_class=tokenizer,
)

print(f"🚀 학습 시작! (총 {len(raw_data)}개 데이터, epoch {EPOCHS})")
trainer.train()
print("✅ 학습 완료!")
```
